The code reads an image file named 'L033_p1.jpg' in grayscale mode using OpenCV's cv.imread function with the flag cv.IMREAD_GRAYSCALE.

## **Image Filtering:**
A median filter with a kernel size of 5x5 (cv.medianBlur(img, 5)) is applied to the grayscale image. This helps reduce noise and smoothens the image.
## **Thresholding Operations:**
Global Thresholding: cv.threshold is applied to the preprocessed image (img). Pixels with intensity values greater than 127 are set to 255, while others are set to 0, creating a binary image (th1).
Adaptive Mean Thresholding: cv.adaptiveThreshold calculates the threshold for smaller regions of the image. It computes the threshold value for each pixel based on the mean of the neighborhood area, creating th2.
Adaptive Gaussian Thresholding: Similar to adaptive mean thresholding, but it uses the weighted sum of neighborhood values, where weights are given by a Gaussian window. The result is stored in th3.

## thresholding Function:
This function takes an input image (image) as a parameter.
Converts the input image from BGR to grayscale using cv2.cvtColor(image, cv2.COLOR_BGR2GRAY).
Applies a thresholding operation on the grayscale image using cv2.threshold.
In this case, the thresholding is set to convert pixel values below 80 to maximum value 255 (white) and above 80 to 0 (black) due to the flag cv2.THRESH_BINARY_INV, creating a binary inverted image (thresh).
Displays the thresholded image using plt.imshow(thresh, cmap='gray') with a grayscale colormap.

```python
def thresholding(image):
    img_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    ret,thresh = cv2.threshold(img_gray,80,255,cv2.THRESH_BINARY_INV)
    plt.imshow(thresh, cmap='gray')
    return thresh

thresh_img = thresholding(img);
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt


def implt(img, cmp=None, t=''):
    """Show image using plt."""
    plt.imshow(img, cmap=cmp)
    plt.title(t)
    plt.show()

def page_detection(image, area_thresh = 0.5):
    """Finding Page."""
    small = resize(image)
    # Edge detection
    image_edges = _edges_detection(small, 200, 250)
    
    # Close gaps between edges (double page clouse => rectangle kernel)
    closed_edges = cv2.morphologyEx(image_edges, 
                                    cv2.MORPH_CLOSE, 
                                    np.ones((5, 11)))
    # Countours
    page_contour = _find_page_contours(closed_edges, small, area_thresh)

    # Recalculate to original scale
    page_contour = page_contour.dot(ratio(image, small.shape[0]))
    # Transform prespective
    new_image = _persp_transform(image, page_contour)
    return new_image
```

```python
def words_detection(image, join=False):
    """Detecting the words bounding boxes.
    Return: numpy array of bounding boxes [x, y, x+w, y+h]
    """
    # Preprocess image for word detection
    blurred = cv2.GaussianBlur(image, (5, 5), 18)
    edge_img = _edge_detect(blurred)
    ret, edge_img = cv2.threshold(edge_img, 50, 255, cv2.THRESH_BINARY)
    bw_img = cv2.morphologyEx(edge_img, cv2.MORPH_CLOSE,
                              np.ones((15,15), np.uint8))

    return _text_detect(bw_img, image, join)
```

```python
SMALL_HEIGHT = 800
def resize(img, height=SMALL_HEIGHT, always=False):
    """Resize image to given height."""
    if (img.shape[0] > height or always):
        rat = height / img.shape[0]
        return cv2.resize(img, (int(rat * img.shape[1]), height))
    
    return img
```

```python
def _edges_detection(img, minVal, maxVal):
    """Preprocessing (gray, thresh, filter, border) + Canny edge detection."""
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    img = cv2.bilateralFilter(img, 9, 75, 75)
    img = cv2.adaptiveThreshold(img, 255,
                                cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                cv2.THRESH_BINARY, 115, 4)

    # Median blur replace center pixel by median of pixels under kelner
    # => removes thin details
    img = cv2.medianBlur(img, 11)

    # Add black border - detection of border touching pages
    img = cv2.copyMakeBorder(img, 5, 5, 5, 5,
                             cv2.BORDER_CONSTANT,
                             value=[0, 0, 0])
    return cv2.Canny(img, minVal, maxVal)
```

```python
def _find_page_contours(edges, img, area_thresh):
    """Finding corner points of page contour."""
    contours, hierarchy = cv2.findContours(edges,
                                                cv2.RETR_TREE,
                                                cv2.CHAIN_APPROX_SIMPLE)
    
    # Finding biggest rectangle otherwise return original corners
    height = edges.shape[0]
    width = edges.shape[1]
    MIN_COUNTOUR_AREA = height * width * area_thresh
    MAX_COUNTOUR_AREA = (width - 10) * (height - 10)

    max_area = MIN_COUNTOUR_AREA
    page_contour = np.array([[0, 0],
                             [0, height-5],
                             [width-5, height-5],
                             [width-5, 0]])

    for cnt in contours:
        perimeter = cv2.arcLength(cnt, True)
        approx = cv2.approxPolyDP(cnt, 0.03 * perimeter, True)

        # Page has 4 corners and it is convex
        if (len(approx) == 4 and
                cv2.isContourConvex(approx) and
                max_area < cv2.contourArea(approx) < MAX_COUNTOUR_AREA):
            
            max_area = cv2.contourArea(approx)
            page_contour = approx[:, 0]

    # Sort corners and offset them
    page_contour = _four_corners_sort(page_contour)
    return _contour_offset(page_contour, (-5, -5))
```

```python
def _four_corners_sort(pts):
    """Sort corners in order: top-left, bot-left, bot-right, top-right."""
    diff = np.diff(pts, axis=1)
    summ = pts.sum(axis=1)
    return np.array([pts[np.argmin(summ)],
                     pts[np.argmax(diff)],
                     pts[np.argmax(summ)],
                     pts[np.argmin(diff)]])
```

```python
def _contour_offset(cnt, offset):
    """Offset contour because of 5px border."""
    cnt += offset
    cnt[cnt < 0] = 0
    return cnt


def ratio(img, height=SMALL_HEIGHT):
    """Getting scale ratio."""
    return img.shape[0] / height

```

```python
def _persp_transform(img, s_points):
    """Transform perspective from start points to target points."""
    # Euclidean distance - calculate maximum height and width
    height = max(np.linalg.norm(s_points[0] - s_points[1]),
                 np.linalg.norm(s_points[2] - s_points[3]))
    width = max(np.linalg.norm(s_points[1] - s_points[2]),
                 np.linalg.norm(s_points[3] - s_points[0]))
    
    # Create target points
    t_points = np.array([[0, 0],
                        [0, height],
                        [width, height],
                        [width, 0]], np.float32)
    
    # getPerspectiveTransform() needs float32
    if s_points.dtype != np.float32:
        s_points = s_points.astype(np.float32)
    
    M = cv2.getPerspectiveTransform(s_points, t_points) 
    return cv2.warpPerspective(img, M, (int(width), int(height)))
```

```python
def _edge_detect(im):
    """ 
    Edge detection using sobel operator on each layer individually.
    Sobel operator is applied for each image layer (RGB)
    """
    return np.max(np.array([_sobel_detect(im[:,:, 0]),
                            _sobel_detect(im[:,:, 1]),
                            _sobel_detect(im[:,:, 2])]), axis=0)

def _sobel_detect(channel):
    """Sobel operator."""
    sobelX = cv2.Sobel(channel, cv2.CV_16S, 1, 0)
    sobelY = cv2.Sobel(channel, cv2.CV_16S, 0, 1)
    sobel = np.hypot(sobelX, sobelY)
    sobel[sobel > 255] = 255
    return np.uint8(sobel)
```

```python
def _text_detect(img, image, join=False):
    """Text detection using contours."""
    small = resize(img, 2000)
    
    # Finding contours
    mask = np.zeros(small.shape, np.uint8)
    cnt, hierarchy = cv2.findContours(np.copy(small),
                                           cv2.RETR_CCOMP,
                                           cv2.CHAIN_APPROX_SIMPLE)
    
    index = 0    
    boxes = []
    # Go through all contours in top level
    while (index >= 0):
        x,y,w,h = cv2.boundingRect(cnt[index])
        cv2.drawContours(mask, cnt, index, (255, 255, 255), cv2.FILLED)
        maskROI = mask[y:y+h, x:x+w]
        # Ratio of white pixels to area of bounding rectangle
        r = cv2.countNonZero(maskROI) / (w * h)
        
        # Limits for text
        if (r > 0.1
            and 1600 > w > 10
            and 1600 > h > 10
            and h/w < 3
            and w/h < 10
            and (60 // h) * w < 1000):
            boxes += [[x, y, w, h]]
            
        index = hierarchy[0][index][0]

    if join:
        # Need more work
        boxes = _group_rectangles(boxes)

    # image for drawing bounding boxes
    small = cv2.cvtColor(small, cv2.COLOR_GRAY2RGB)
    bounding_boxes = np.array([0,0,0,0])
    for (x, y, w, h) in boxes:
        cv2.rectangle(small, (x, y),(x+w,y+h), (0, 255, 0), 2)
        bounding_boxes = np.vstack((bounding_boxes,
                                    np.array([x, y, x+w, y+h])))
        
    implt(small, t='Bounding rectangles')
    
    boxes = bounding_boxes.dot(ratio(image, small.shape[0])).astype(np.int64)
    return boxes[1:]  
```

```python
import tensorflow.compat.v1 as tf
from tensorflow.python.ops.rnn_cell_impl import LSTMCell, ResidualWrapper, DropoutWrapper, MultiRNNCell

class Model():
    """Loading and running isolated tf graph."""
    def __init__(self, loc, operation='activation', input_name='x'):
        """
        loc: location of file containing saved model
        operation: name of operation for running the model
        input_name: name of input placeholder
        """
        self.input = input_name + ":0"
        self.graph = tf.Graph()
        self.sess = tf.Session(graph=self.graph)
        with self.graph.as_default():
            saver = tf.train.import_meta_graph(loc + '.meta', clear_devices=True)
            saver.restore(self.sess, loc)
            self.op = self.graph.get_operation_by_name(operation).outputs[0]

    def run(self, data):
        """Run the specified operation on given data."""
        return self.sess.run(self.op, feed_dict={self.input: data})
    
    def eval_feed(self, feed):
        """Run the specified operation with given feed."""
        return self.sess.run(self.op, feed_dict=feed)
    
    def run_op(self, op, feed, output=True):
        """Run given operation with the feed."""
        if output:
            return self.sess.run(
                self.graph.get_operation_by_name(op).outputs[0],
                feed_dict=feed)
        else:
            self.sess.run(
                self.graph.get_operation_by_name(op),
                feed_dict=feed)
        
    
    
def _create_single_cell(cell_fn, num_units, is_residual=False, is_dropout=False, keep_prob=None):
    """Create single RNN cell based on cell_fn."""
    cell = cell_fn(num_units)
    if is_dropout:
        cell = DropoutWrapper(cell, input_keep_prob=keep_prob)
    if is_residual:
        cell = ResidualWrapper(cell)
    return cell


def create_cell(num_units, num_layers, num_residual_layers, is_dropout=False, keep_prob=None, cell_fn=LSTMCell):
    """Create corresponding number of RNN cells with given wrappers."""
    cell_list = []
    
    for i in range(num_layers):
        cell_list.append(_create_single_cell(
            cell_fn=cell_fn,
            num_units=num_units,
            is_residual=(i >= num_layers - num_residual_layers),
            is_dropout=is_dropout,
            keep_prob=keep_prob
        ))

    if num_layers == 1:
        return cell_list[0]
    return MultiRNNCell(cell_list)
```

## Analyzing handwriting style attributes involves extracting various features like slant, size, curvature, spacing, pressure, etc., from handwritten text images and using these features to infer characteristics about the writer.

## Reads a handwritten text image and extracts its features related to slant, size, and curvature.
## Calculates the slant angle by fitting an ellipse to each contour and measuring its orientation.
## Computes the size of each contour (area) and calculates a curvature measure (ratio of contour perimeter to area).

```python
import cv2
import numpy as np

# Read the handwritten text image
image = cv2.imread('/Users/tigrangaplanyan/Downloads/ImageJ/Midterm_Project/L033/L033_p1_handwritten.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Thresholding to obtain a binary image
_, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Invert the binary image
binary = 255 - binary

# Find contours
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Calculate slant, size, and curvature-related features
slant_angles = []
sizes = []
curvatures = []

for contour in contours:
    if cv2.contourArea(contour) > 50:  # Filter out small contours (adjust threshold as needed)
        # Fit an ellipse to each contour
        ellipse = cv2.fitEllipse(contour)
        (x, y), (MA, ma), angle = ellipse

        # Slant angle calculation
        slant_angle = 90.0 - angle if angle > 90 else angle - 90.0
        slant_angles.append(slant_angle)

        # Size calculation
        size = cv2.contourArea(contour)
        sizes.append(size)

        # Curvature calculation (example: ratio of contour perimeter to area)
        perimeter = cv2.arcLength(contour, True)
        if size > 0:
            curvature = perimeter / size
            curvatures.append(curvature)

# Analyze extracted features (for demonstration purposes)
mean_slant_angle = np.mean(slant_angles)
mean_size = np.mean(sizes)
mean_curvature = np.mean(curvatures)

print("Analyzing Handwriting Style Attributes:")
print(f"Mean Slant Angle: {mean_slant_angle:.2f} degrees")
print(f"Mean Size: {mean_size:.2f} pixels")
print(f"Mean Curvature: {mean_curvature:.2f}")
```

## This code generates a simple fractal-like curve and estimates its fractal dimension using the box-counting method.
## The technique of calculating the fractal dimension of handwritten text or its contours is one approach among many for analyzing the complexity or irregularity of shapes in the text strokes.

## **Points and Applications of Fractal Dimension Analysis in Handwritten Text:**

## **Complexity Analysis:** Fractal dimension estimation helps quantify the complexity or irregularity of shapes within the handwritten text strokes.

## **Differentiating Writing Styles:** It may differentiate between different writing styles based on their inherent complexities or irregularities. For instance, comparing fractal dimensions across different writers might reveal distinctive stylistic differences.

## **Psychological Inferences:** Some studies suggest correlations between handwriting characteristics and psychological traits. Analyzing fractal dimensions might attempt to infer certain psychological traits or tendencies from handwriting complexities, though this is a highly debated area and not definitively established.

## **Forensic Analysis:** In forensic handwriting analysis, examining the complexity of strokes might aid in identifying forgeries or determining if multiple samples were written by the same person.

## **Data Augmentation:** In machine learning applications for handwriting recognition, understanding the complexity of text strokes might assist in generating synthetic data or augmenting existing datasets for better training of recognition models.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the handwritten text image
image = cv2.imread('/Users/tigrangaplanyan/Downloads/ImageJ/Midterm_Project/L033/L033_p1_handwritten.jpg', cv2.IMREAD_GRAYSCALE)

# Thresholding to obtain a binary image
_, binary = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Find contours
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Function to calculate fractal dimension using box-counting method
def calculate_fractal_dimension(contour):
    contour_perimeter = cv2.arcLength(contour, True)
    contour_area = cv2.contourArea(contour)
    fractal_dimension = 2 * np.log(contour_perimeter / (4 * np.sqrt(np.pi * contour_area)))
    return fractal_dimension

# Calculate fractal dimensions for all contours
contour_dimensions = []
for contour in contours:
    if cv2.contourArea(contour) > 50:  # Filter out small contours (adjust threshold as needed)
        fractal_dimension = calculate_fractal_dimension(contour)
        contour_dimensions.append(fractal_dimension)

# Display calculated fractal dimensions
print("Fractal Dimensions of Contours:")
print(contour_dimensions)

# Display a histogram of calculated fractal dimensions
plt.hist(contour_dimensions, bins=20, edgecolor='black')
plt.xlabel('Fractal Dimension')
plt.ylabel('Frequency')
plt.title('Histogram of Fractal Dimensions')
plt.show()
```

## This code snippet performs the following operations:

## Converts the image to grayscale.
## Applies adaptive thresholding to obtain a binary image.
## Uses morphological operations (closing) to enhance text regions.
## Finds contours of potential text regions and draws rectangles around those areas.

```python
import cv2
import numpy as np

# Read the image with handwritten text
image_path = '/Users/tigrangaplanyan/Downloads/ImageJ/Midterm_Project/L033/L033_p1_handwritten.jpg'
image = cv2.imread(image_path)

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply adaptive thresholding to obtain a binary image
binary = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2)

# Perform morphological operations to enhance text regions
kernel = np.ones((3, 3), np.uint8)
binary = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel, iterations=2)

# Find contours of potential text regions
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Iterate through contours to filter out smaller regions and draw rectangles around the text-like areas
for contour in contours:
    area = cv2.contourArea(contour)
    if area > 1000:  # Adjust area threshold based on text size
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Display the image with detected potential text regions
cv2.imshow('Detected Text Regions', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## This code reads an image with handwritten text, converts it to grayscale, and applies thresholding to obtain a binary image.
## It identifies potential text regions using contours and calculates the area of each contour (representing characters).

## This code reads an image with handwritten text, converts it to grayscale, and applies thresholding to obtain a binary image.

## It identifies potential text regions using contours and calculates the bounding rectangles for each contour (representing characters).

## The code calculates the x-coordinate differences between consecutive character rectangles to estimate the average spacing between characters.


```python
import cv2
import numpy as np

# Read the image with handwritten text
image_path = '/Users/tigrangaplanyan/Downloads/ImageJ/Midterm_Project/L033/L033_p1_handwritten.jpg'
image = cv2.imread(image_path)

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding to obtain a binary image
_, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Find contours to identify potential text regions
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Calculate the bounding rectangles of each contour (character)
character_rects = [cv2.boundingRect(contour) for contour in contours]

# Calculate the average spacing between characters
if len(character_rects) > 1:
    character_rects.sort(key=lambda x: x[0])  # Sort rectangles by x-coordinate
    spacings = [character_rects[i+1][0] - (character_rects[i][0] + character_rects[i][2]) for i in range(len(character_rects)-1)]
    average_spacing = sum(spacings) / len(spacings)
    print(f"Average Spacing Between Characters: {average_spacing:.2f} pixels")
else:
    print("Insufficient number of characters to calculate spacing.")
```

## This code reads an image with handwritten text and converts it to grayscale.
## It applies a Gaussian blur to reduce noise and calculates the Laplacian to detect edges or areas with high gradients (representing strokes).

## The Laplacian values are converted to absolute values, and thresholding is applied to obtain a binary image highlighting edges.

## The code calculates the variance of stroke thickness in the binary edge-detected image.

```python
import cv2
import numpy as np

# Read the image with handwritten text
image_path = '/Users/tigrangaplanyan/Downloads/ImageJ/Midterm_Project/L033/L033_p1_handwritten.jpg'
image = cv2.imread(image_path)

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Calculate the Laplacian to detect edges
laplacian = cv2.Laplacian(blurred, cv2.CV_64F)

# Convert Laplacian to absolute values
laplacian = np.uint8(np.absolute(laplacian))

# Apply thresholding to obtain a binary image highlighting edges
_, edges_binary = cv2.threshold(laplacian, 30, 255, cv2.THRESH_BINARY)

# Calculate the variance of stroke thickness
variance = np.var(edges_binary)

print(f"Variance of Stroke Thickness: {variance:.2f}")
```

## This code reads an image with handwritten text, converts it to grayscale, and applies thresholding to obtain a binary image.
## It identifies potential text regions using contours and calculates the baseline shift for each character by measuring the difference between the bottom of the character's bounding rectangle and its baseline.

```python
import cv2
import numpy as np

# Read the image with handwritten text
image_path = '/Users/tigrangaplanyan/Downloads/ImageJ/Midterm_Project/L033/L033_p1_handwritten.jpg'
image = cv2.imread(image_path)

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding to obtain a binary image
_, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Invert the binary image
binary = cv2.bitwise_not(binary)

# Find contours to identify potential text regions
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Calculate the baseline shift of characters
baseline_shifts = []
for contour in contours:
    # Calculate the bounding rectangle of the contour
    x, y, w, h = cv2.boundingRect(contour)
    
    # Calculate the baseline shift as the difference between the bottom of the bounding rectangle and the baseline
    baseline_shift = y + h - (y + h // 2)
    baseline_shifts.append(baseline_shift)

# Calculate the average baseline shift of characters
if baseline_shifts:
    average_baseline_shift = sum(baseline_shifts) / len(baseline_shifts)
    print(f"Average Baseline Shift of Characters: {average_baseline_shift:.2f} pixels")
else:
    print("No handwritten text regions found.")
```
